# PHASE 2: BUSINESS LOGIC (FILTERING COMPETITION / HOSTELERIA)

### PHASE 2.A: FILTERING COMPETITION / HOSTELERIA

```python
print("🎯 Filtering dataset to isolate the Hospitality and Restaurant sector...\n")

# 1. Let's see the unique high-level sectors available to confirm the exact name
available_sectors = df_census['Nom_Grup_Activitat'].unique()
print(f"📋 Available groups in Barcelona:\n{available_sectors}\n")
```

```python
# 2. We use the exact group name discovered
keywords_group = 'Restaurants, bars i hotels'
keywords_activity = 'Restaurants|Bars|CIBERCAFÈ|Degustació|Xocolateries|Geladeries|Bars especials amb actuació|Bars musicals|Discoteques|PUB'

# 3. Apply the surgical filter
df_competitors = df_census[
    df_census['Nom_Grup_Activitat'].str.contains(keywords_group, case=False, na=False) |
    df_census['Nom_Activitat'].str.contains(keywords_activity, case=False, na=False)
].copy()

# 4. Calculate data reduction
initial_stores = df_census.shape[0]
filtered_stores = df_competitors.shape[0]
reduction_pct = (filtered_stores / initial_stores) * 100

print("=" * 60)
print(f"📉 Initial commercial locations: {initial_stores:,}")
print(f"🔥 Active Hospitality and Restaurant competitors found: {filtered_stores:,}")
print(f"💹 The target market represents {reduction_pct:.2f}% of the total city census.")
```

### PHASE 2.B: DATA QUALITY (NULL AUDIT & CLEANING)

```python
print("🔍 Auditing critical missing values of our Competitors...\n")

# 1. Define the columns we absolutely need to be clean
critical_cols = ['Codi_Districte', 'Nom_Districte', 'Latitud', 'Longitud', 'Nom_Activitat']

# 2. Check for nulls only in these columns
nulls_before = df_competitors[critical_cols].isnull().sum()
print("⚠️ Nulls detected before cleaning:")
print(nulls_before[nulls_before > 0] if nulls_before.sum() > 0 else "✅ No nulls found!")
print("=" * 60)

# 3. Drop rows that are missing District or Coordinates (Ghost locations)
df_competitors = df_competitors.dropna(subset=['Codi_Districte', 'Latitud', 'Longitud'])

# 4. Final check
print(f"🧹 Clean dataset size: {df_competitors.shape[0]:,} valid competitors ready for spatial analysis.")
```

# PHASE 3: SPATIAL AGGREGATION (SUPPLY SATURATION BY DISTRICT)

```python
print("📊 Aggregating active competitors by District to measure market saturation...\n")

# 1. Group by District Code and Name, then count the number of establishments
df_supply = df_competitors.groupby(
    ['Codi_Districte', 'Nom_Districte']
).size().reset_index(name='total_competitors')

# 2. Sort the districts by the highest number of competitors (highest saturation)
df_supply = df_supply.sort_values(by='total_competitors', ascending=False).reset_index(drop=True)

# 3. Calculate the percentage of total competition each district holds
total_bcn_competitors = df_supply['total_competitors'].sum()
df_supply['saturation_pct'] = (df_supply['total_competitors'] / total_bcn_competitors) * 100

# 4. Display the results
print("🏆 Top Districts by Hospitality and Restaurant Saturation (Supply):")
display(df_supply.style.format({
    'saturation_pct': '{:.2f}%',
    'total_competitors': '{:,}'
}))
```

# PHASE 4: DATA EXPORT (PIPELINE HANDOFF)

```python
print("💾 Saving the cleaned censu_bcn dataset...")

# 1. Directory defined in config.py
processed_dir = config.PROCESSED_DATA_DIR

# 2. Ensure the directory exists (it creates it if you haven't yet)
os.makedirs(processed_dir, exist_ok=True)

# 3. Save the detailed competitors map (11,097 rows)
output_path1 = os.path.join(processed_dir, "02A_bcn_competitors_cleaned.csv")

# 4. Save the grouped saturation table (10 rows)
output_path2 = os.path.join(processed_dir, "02B_bcn_district_saturation.csv")

# 4. Save the DataFrame to CSV without the pandas row index
df_competitors.to_csv(output_path1, index=False)
df_supply.to_csv(output_path2, index=False)

print(f"✅ Success! The 'Supply & Competition' data is safely stored at:")
print(f"📂 {output_path1}")
print(f"📂 {output_path2}")
print("\n🎉 EDA CENSU_BCN COMPLETE.")
```

```python
# ==============================================================================
# SPATIAL VISUALIZATION (HEATMAP)
# ==============================================================================

import folium
from folium.plugins import HeatMap

print("🗺️ Generating the interactive HeatMap of Barcelona...")

# 1. Create the map "canvas" centered on Barcelona's general coordinates
mapa_bcn = folium.Map(location=[41.3874, 2.1686], zoom_start=12, tiles='cartodbpositron')

# 2. Extract the coordinates of our 11,097 clean competitors
# Convert them into a mathematical list that the map can read
coordenadas = df_competitors[['Latitud', 'Longitud']].values.tolist()

# 3. Add the heat layer to the map (adjusting the radius and blur of the spots)
HeatMap(coordenadas, radius=8, blur=5).add_to(mapa_bcn)

# 4. Save the map as an independent HTML file
html_path = os.path.join(processed_dir, "02C_bcn_heatmap.html")

mapa_bcn.save(html_path)

print(f"✅ Success! The interactive HeatMap is safely stored at:")
print(f"📂 {html_path}")
```
